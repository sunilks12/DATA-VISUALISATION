# IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Load the Data

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

# BARPLOT

```python
sns.barplot(x='sex', y='total_bill', data=tips)
plt.show()
```

# BOXPLOT

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='smoker',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='smoker',y='total_bill',hue='sex',data=tips,palette = 'Pastel2')
plt.show()
```

# COUNTPLOT

```python
sns.countplot(x='day',data=tips)
plt.show()
```

```python
sns.countplot(x='time',data=tips)
plt.show()
```

```python
sns.countplot(x='time',hue='smoker',data=tips)
plt.show()
```

# VIOLIN PLOT

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips)
plt.show()
```

```python
#if we keep split we will get male and female combined
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,split=True)
plt.show()
```

# JITTER PLOT

```python
#if we keep jitter dots will come spacely
sns.stripplot(x='day',y='total_bill',hue='sex',jitter = True, data=tips)
plt.show()
```

# STRIPPLOT

```python
#if we keep dodge in stripplot we will get smoker different columns
sns.stripplot(x='day',y='total_bill',hue='smoker',jitter = True, data=tips,dodge=True)
plt.show()
```

# SWARMPLOT

```python
#it gives some shape 
sns.swarmplot(x='day',y='total_bill',hue='sex',dodge=True,data=tips)
plt.show()
```

# DISTRIBUTION PLOTS

```python
import matplotlib as mpl
```

```python
#kernel density estimation(kde) it will give probabilty if it is true or it will give the frequency
mpl.rcParams['patch.force_edgecolor']=True
mpl.rcParams['patch.edgecolor']='red'
sns.distplot(tips['total_bill'],kde=False)
plt.title('Distribution Plot')
plt.ylabel('probabaility')
plt.show()
```

```python
#we can see correlation in this plot and it is positive skewed by default it is scatter 
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#if we put kind as reg we will get regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind ='reg')
plt.show()
```

```python
# if we keep kernel density we will get density
sns.jointplot(x='total_bill',y='tip',data=tips,kind ='kde')
plt.show()
```

```python
#if we want to get the relation between numerical columns we use pairplot
sns.pairplot(tips)
```
